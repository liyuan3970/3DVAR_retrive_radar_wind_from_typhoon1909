二、第二审稿人审稿专家初审意见：

台风强降水多尺度特征和发生机理是热带气旋研究的重点和难点，本文对“利奇马”（2019）引发强降雨的中尺度特征进行分析，选题有意义。但本文存在如下明显不足：
1. 本文的题目不切题。本文研究的降水为台风登陆期间的本体降水，降水的中尺度分布特征与台风自身结构和浙中南小尺度地形关系密切，因此本文研究目标是强降水的中尺度特征和发生机理而非仅仅是风场特征分析。
2. 本文引言部分文献索引的针对性和逻辑性不强，文献调研不充分。本文的文献调研没有针对本文主旨“台风强降水的中尺度特征及成因”开展充分调研，调研的广度和深度都不足；行文表达中，更多是泛泛罗列，文献综述整体缺乏学术逻辑性。
3. 文中所用资料和方法描述过于简单，对使用资料的可靠性缺乏充分检验，对关键论据论证不够充分和深入，所得结论可靠性不足。本文要分析中尺度特征，高分辨的可靠资料是关键，文中尝试了应用多种数据，但这些数据的可靠性评估检验不足，关键问题分析缺乏足够证据支撑。比如：对于雷达反演风场方法介绍不足，对于于EC资料做的反演风场检验，仅使用EC资料检验存在资料同源之嫌，应增加CRA或者NCEP等其他再分析资料的对比评估。又如：本文开展了中尺度数值模拟，模式设计和配置的情况说明不足，对于模拟结果的检验客观定量不足，模拟资料的可靠性存疑，这些不足均直接影响到分析结论的可靠性。还有：对于地形重力波的分析中，能够产生重力波不等于实际中一定激发了重力波，文中未提供地形重力波关键动力特征的直接证据（波速，波长等）？由此相关分析得到的结论显得力不从心。
综上所述，本文研究选题有意义，但论证分析不够充分，所得结论可靠性存疑，参考价值不大。建议退稿。


三、第三审稿人审稿专家初审意见：

本文对1909号超强台风“利奇马”登陆浙江引发强降雨的中尺度流场特征进行了分析，对两个区域强降水的形成原因进行了基于观测和数值模拟的分析，发现雁荡山的降水主要是与中尺度复合线北抬减弱有关，括苍山的降水主要与偏北急流上坡引发的重力波上升运动有关。文章有一定新颖性，特别是在揭示中尺度系统对强降水影响方面，但是整篇文章的分析还比较粗糙，如果能再细致一些将是一篇很不错的文章。评阅建议如下：

1）关于地形重力波的诊断部分，吴等人的研究表明，当大气层结不稳定时，在坡度增加方向上将有利于重力波的发展，并非是作者论述的“当气流与山地走向一致时能够在山前产生重力波”，在山前还是背风产生重力波波动取决于大气的层结不稳定状态，这需要从相当位温的垂直分布分析。从作者的图中可以清晰看到重力波，如图3f和图5b是相互印证的，但是作者的相关理解或者论述不太科学，建议结合理论以及你自己的结果进行重新梳理。

2）本文虽然主要针对的是中尺度流场的分析，但是任何中尺度环境离不开天气尺度流场背景，特别是本文发现雁荡山中尺度辐合线移动缓慢，为什么为缓慢，原因是什么，应该和大尺度流场有关；另外，对括苍山，作者说北部的偏北急流非常重要，它冲向地形产生重力波，那么这个偏北急流是如何产生的，为何生成在这个位置？在该急流的其他位置无地形处，是否也有类似的波动，以此可以证明是否是地形激发产生。另外，作者做了数值模拟，也可以验证偏北急流的强弱、地形的高低、以及上干下湿的环境是否真的导致山前重力波的产生。在利奇马登陆前，天气背景流场上显示西太副高、大陆高压脊和西风槽都比较明显，这样的大尺度流场对中尺度系统的孕育和维持起了不可或缺的作用，建议作者结合起来一并分析。

3）文章的标题“ 1909号超强台风“利奇马”引发登陆强降雨的中尺度风场特征分析”需要修改，这个标题看着更像是仅针对大风的中尺度分析，但数值模拟的部分是综合的，文章目的是对台风登陆引发的强降水进行分析，所以标题变成“1909号超强台风“利奇马”登陆引发浙江强降雨的中尺度特征分析”或者类似于这样的表达更合适。

4）作者描述山前中尺度流场特征（图5）时主要用的是某一点的剖面，单点的描述存在片面性，建议用区域平均更合理。

5）作者提到括苍山前存在低空不稳定，同时由RH可见那里的相对湿度很大，因此在图5中的位温应该换成相当位温，而不是位温。

6）图题描述不清，如图3d中的两条曲线的意义、图3、图5各物理量的单位、实线、虚线意义等。

7）表1中，03和05点的风向是否有误？为什么是32，是不是320？需核实。此外，对风向应该有一个基本的定义或描述，如东风是多少度，北风是多少度等，增强可读性。

8）图的可读性比较差，如图2的几个点可以和地形图叠加显示，这样作者介绍几个点位置时（如平原山区交界等）更加一目了然；图1重点分析的1和2两个区域，可以同时在图3和图4中叠加；有的图不用拼在一起，不然很小，除非版面有限，这样读起来和分析起来都比较清晰。

9）英文摘要不少错误，包括词汇，语法，标点，格式等等。
