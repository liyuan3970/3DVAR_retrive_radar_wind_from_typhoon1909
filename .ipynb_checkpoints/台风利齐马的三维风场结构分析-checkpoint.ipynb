# 台风利齐马的三维风场结构分析

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;众多观测事实表明，地形对台风降水的作用非常明显[5]而台风自身环流中蕴含丰富的水汽，和急流，当气流遇到山地时，有利于激发对流，进而在山区产生强降水。为了研究“利奇马”台风登陆前后对强降水成因，本节统计台风登陆前后8月9日23:00到10日05点的降水分布如下图：图中，散点图为浙江省中尺度自动站和国家站降水分布，填充色为地形海拔高度，受此次台风降水影响较大地区主要集中浙江东部宁波、台州、杭州、嘉兴等沿海地区，约500个站累积降水超过(500)毫米,沿海地区整体地势较为平坦，宁波以北地区以平原为主，而平均海拔高于500米的山区主要集中在浙江中部温台交界，台风于10日凌晨1点47分登陆台州温岭石塘，由于温台山区距离台风较近，降水非常明显，全省500毫米以上强降水站几乎都分布于此，约200站（黄色）降水主要分布于括苍山脉和雁荡山脉附近，雨强大于500毫米的站主要集中在温台交界以及括苍山脉附近（红色），其中温台交界处降水相对集中，有50站超过500毫米，而最大累积降水在括苍山站，达340毫米。该地最高1383m，区域平均海拔1000米以上，可见地形对降水的增幅作用非常明显。地形常常能够为提高降水效率提供动力机制，之后本文将着重分析强降水中心形成以及降水极大值出现在括苍山的原因
